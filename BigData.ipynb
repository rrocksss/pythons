```python
#import libraries
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
import numpy as np

data = pd.read_csv('/content/sample_data/dataset.csv')
data.columns = data.columns.str.replace(' ','_')
```

```python
#print data by subplot for each label/feature
i = 1
plt.figure(figsize = (15,25))
for feature in data:
    plt.subplot(5,3,i)
    sns.countplot(x = feature, data = data)
    i +=1
```

```python
#print data by subplot for each label/feature according to adaptivity level
i = 1
plt.figure(figsize = (15,25))
for feature in data:
    plt.subplot(5,3,i)
    sns.countplot(x = feature ,  data = data, hue = 'Adaptivity_Level')
    i +=1
```

```python
#print data by subplot for each label/feature that adaptivity level is low
LowLevelData = data[data['Adaptivity_Level'] == "Low"]
i = 1
plt.figure(figsize = (15,25))
for feature in data:
    plt.subplot(5,3,i)
    sns.countplot(x = feature , data = LowLevelData)
    i +=1
```

```python
#Find percentage that effects online studying effectiveness and visualize in bar chart

#Gender
Prob_Boy = LowLevelData.Gender.value_counts().Boy * 100/ data.Gender.value_counts().Boy
Prob_Girl = LowLevelData.Gender.value_counts().Girl * 100/ data.Gender.value_counts().Girl
print("Gender:")
print("Boy:")
print( Prob_Boy)
print("Girl:")
print(Prob_Girl)
print("Diff:")
print(max(Prob_Boy, Prob_Girl) - min(Prob_Boy, Prob_Girl))
y = np.array([Prob_Boy, Prob_Girl])
mylabels = ["%Boys_Low_Adaptivity", "%Girls_Low_Adaptivity"]
plt.figure(figsize = (5,5))
plt.bar(mylabels, y)
plt.show()
print("\n")
print("\n")


#Education_Level
Prob_School = LowLevelData.Education_Level.value_counts().School*100/ data.Education_Level.value_counts().School
Prob_College = LowLevelData.Education_Level.value_counts().College *100/ data.Education_Level.value_counts().College
Prob_University = LowLevelData.Education_Level.value_counts().University *100/ data.Education_Level.value_counts().University
print("Education_Level:")
print("School:")
print( Prob_School)
print("College:")
print(Prob_College)
print("University:")
print(Prob_University)
print("Diff:")
print(max(Prob_School, Prob_College, Prob_University) - min(Prob_School, Prob_College, Prob_University))
y = np.array([Prob_School, Prob_College, Prob_University])
mylabels = ["%School_Low_Adaptivity", "%College_Low_Adaptivity", "%University_Low_Adaptivity"]
plt.figure(figsize = (8,5))
plt.bar(mylabels, y)
plt.show()
print("\n")
print("\n")

#Institution_Type
Prob_Govern = LowLevelData.Institution_Type.value_counts().Government *100/ data.Institution_Type.value_counts().Government
Prob_not_Govern = LowLevelData.Institution_Type.value_counts()['Non Government'] *100/ data.Institution_Type.value_counts()['Non Government']
print("Institution_Type:")
print("Government:")
print( Prob_Govern)
print("not_Govern:")
print(Prob_not_Govern)
print("Diff:")
print(max(Prob_Govern, Prob_not_Govern) - min(Prob_Govern, Prob_not_Govern))
y = np.array([Prob_Govern, Prob_not_Govern])
mylabels = ["%Government_Low_Adaptivity", "%not_Government_Low_Adaptivity"]
plt.figure(figsize = (6,5))
plt.bar(mylabels, y)
plt.show()
print("\n")
print("\n")

#IT_Student
Prob_IT = LowLevelData.IT_Student.value_counts().Yes *100/ data.IT_Student.value_counts().Yes
Prob_not_IT = LowLevelData.IT_Student.value_counts().No *100/ data.IT_Student.value_counts().No
print("IT_Student:")
print("IT:")
print( Prob_IT)
print("not_IT:")
print(Prob_not_IT)
print("Diff:")
print(max(Prob_IT, Prob_not_IT) - min(Prob_IT, Prob_not_IT))
y = np.array([Prob_IT, Prob_not_IT])
mylabels = ["%IT_Low_Adaptivity", "%not_IT_Low_Adaptivity"]
plt.figure(figsize = (5,5))
plt.bar(mylabels, y)
plt.show()
print("\n")
print("\n")

#Location
Prob_Town = LowLevelData.Location.value_counts().Yes *100/ data.Location.value_counts().Yes
Prob_not_Town = LowLevelData.Location.value_counts().No *100/ data.Location.value_counts().No
print("Location:")
print("Town:")
print( Prob_Town)
print("not_Town:")
print(Prob_not_Town)
print("Diff:")
print(max(Prob_Town, Prob_not_Town) - min(Prob_Town, Prob_not_Town))
y = np.array([Prob_Town, Prob_not_Town])
mylabels = ["%Town_Low_Adaptivity", "%not_Town_Low_Adaptivity"]
plt.figure(figsize = (5,5))
plt.bar(mylabels, y)
plt.show()
print("\n")
print("\n")

#Load-shedding
Prob_Low = LowLevelData['Load-shedding'].value_counts().Low *100/ data['Load-shedding'].value_counts().Low
Prob_High = LowLevelData['Load-shedding'].value_counts().High *100/ data['Load-shedding'].value_counts().High
print("Load-shedding:")
print("Low:")
print( Prob_Low)
print("High:")
print(Prob_High)
print("Diff:")
print(max(Prob_Low, Prob_High) - min(Prob_Low, Prob_High))
y = np.array([Prob_Low, Prob_High])
mylabels = ["%Low_Load-shedding_Low_Adaptivity", "%High_Load-shedding_Low_Adaptivity"]
plt.figure(figsize = (7,5))
plt.bar(mylabels, y)
plt.show()
print("\n")
print("\n")

#Financial_Condition
Prob_Poor = LowLevelData.Financial_Condition.value_counts().Poor*100/ data.Financial_Condition.value_counts().Poor
Prob_Mid = LowLevelData.Financial_Condition.value_counts().Mid *100/ data.Financial_Condition.value_counts().Mid
Prob_Rich = LowLevelData.Financial_Condition.value_counts().Rich *100/ data.Financial_Condition.value_counts().Rich
print("Financial_Condition:")
print("Poor:")
print( Prob_Poor)
print("Mid:")
print(Prob_Mid)
print("Rich:")
print(Prob_Rich)
print("Diff:")
print(max(Prob_Poor, Prob_Mid, Prob_Rich) - min(Prob_Poor, Prob_Mid, Prob_Rich))
y = np.array([Prob_Poor, Prob_Mid, Prob_Rich])
mylabels = ["%Poor_Low_Adaptivity", "%Mid_Low_Adaptivity", "%Rich_Low_Adaptivity"]
plt.figure(figsize = (8,5))
plt.bar(mylabels, y)
plt.show()
print("\n")
print("\n")

#Internet_Type
Prob_MobileData = LowLevelData.Internet_Type.value_counts()['Mobile Data'] *100/ data.Internet_Type.value_counts()['Mobile Data']
Prob_Wifi = LowLevelData.Internet_Type.value_counts().Wifi *100/ data.Internet_Type.value_counts().Wifi
print("Internet_Type:")
print("Mobile Data:")
print( Prob_MobileData)
print("Wifi:")
print(Prob_Wifi)
print("Diff:")
print(max(Prob_MobileData, Prob_Wifi) - min(Prob_MobileData, Prob_Wifi))
y = np.array([Prob_MobileData, Prob_Wifi])
mylabels = ["%MobileData_Low_Adaptivity", "%Wifi_Low_Adaptivity"]
plt.figure(figsize = (6,5))
plt.bar(mylabels, y)
plt.show()
print("\n")
print("\n")

#Network_Type
Prob_2G = LowLevelData.Network_Type.value_counts()['2G'] *100/ data.Network_Type.value_counts()['2G']
Prob_3G = LowLevelData.Network_Type.value_counts()['3G'] *100/ data.Network_Type.value_counts()['3G']
Prob_4G = LowLevelData.Network_Type.value_counts()['4G'] *100/ data.Network_Type.value_counts()['4G']
print("Network_Type:")
print("2G:")
print( Prob_2G)
print("3G:")
print(Prob_3G)
print("4G:")
print(Prob_4G)
print("Diff:")
print(max(Prob_2G, Prob_3G, Prob_4G) - min(Prob_2G, Prob_3G, Prob_4G))
y = np.array([Prob_2G, Prob_3G, Prob_4G])
mylabels = ["%2G_Low_Adaptivity", "%3G_Low_Adaptivity", "%4G_Low_Adaptivity"]
plt.figure(figsize = (8,5))
plt.bar(mylabels, y)
plt.show()
print("\n")
print("\n")

#Class_Duration
Prob_0 = LowLevelData.Class_Duration.value_counts()['0'] *100/ data.Class_Duration.value_counts()['0']
Prob_1_3 = LowLevelData.Class_Duration.value_counts()['1-3'] *100/ data.Class_Duration.value_counts()['1-3']
Prob_3_6 = LowLevelData.Class_Duration.value_counts()['3-6'] *100/ data.Class_Duration.value_counts()['3-6']
print("Class_Duration:")
print("0:")
print( Prob_0)
print("1-3:")
print(Prob_1_3)
print("3-6:")
print(Prob_3_6)
print("Diff:")
print(max(Prob_0, Prob_1_3, Prob_3_6) - min(Prob_0, Prob_1_3, Prob_3_6))
y = np.array([Prob_0, Prob_1_3, Prob_3_6])
mylabels = ["%0_Low_Adaptivity", "%1-3_Low_Adaptivity", "%3-6_Low_Adaptivity"]
plt.figure(figsize = (8,5))
plt.bar(mylabels, y)
plt.show()
print("\n")
print("\n")

#Self_Lms
Prob_Self = LowLevelData.Self_Lms.value_counts().Yes *100/ data.Self_Lms.value_counts().Yes
Prob_not_Self = LowLevelData.Self_Lms.value_counts().No *100/ data.Self_Lms.value_counts().No
print("Self_Lms:")
print("Yes:")
print( Prob_Self)
print("No:")
print(Prob_not_Self)
print("Diff:")
print(max(Prob_Self, Prob_not_Self) - min(Prob_Self, Prob_not_Self))
y = np.array([Prob_Self, Prob_not_Self])
mylabels = ["%Self_Low_Adaptivity", "%not_Self_Low_Adaptivity"]
plt.figure(figsize = (5,5))
plt.bar(mylabels, y)
plt.show()
print("\n")
print("\n")

#Device
Prob_Mobile = LowLevelData.Device.value_counts().Mobile*100/ data.Device.value_counts().Mobile
Prob_Computer = LowLevelData.Device.value_counts().Computer *100/ data.Device.value_counts().Computer
Prob_Tab = LowLevelData.Device.value_counts().Tab *100/ data.Device.value_counts().Tab
print("Device:")
print("Mobile:")
print( Prob_Mobile)
print("Computer:")
print(Prob_Computer)
print("Tab:")
print(Prob_Tab)
print("Diff:")
print(max(Prob_Mobile, Prob_Computer, Prob_Tab) - min(Prob_Mobile, Prob_Computer, Prob_Tab))
y = np.array([Prob_Mobile, Prob_Computer, Prob_Tab])
mylabels = ["%Mobile_Low_Adaptivity", "%Computer_Low_Adaptivity", "%Tab_Low_Adaptivity"]
plt.figure(figsize = (8,5))
plt.bar(mylabels, y)
plt.show()
print("\n")
print("\n")
```
