```python
# Import libraries yang akan dipakai
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Membaca dataset dan disimpan ke variabel
data = pd.read_csv('/content/sample_data/dataset.csv')
```

```python
# Melihat ada berapa kolom dan datatype apa saja pada data
data.info()
```

```python
# Visualisasi data menjadi grafik scatter plot
plt.scatter(data.iloc[:,0], data.iloc[:,1])
plt.xlabel("x")
plt.ylabel("y")
plt.title("Visualisasi dataset")
plt.show
```

```python
# Import library kmeans dari sklearn.cluster untuk membantu proses clustering
from sklearn.cluster import KMeans
```

```python
# Menyiapkan kmeans untuk membagi data menjadi beberapa cluster, disini saya membagi data menjadi 7 cluster
kmeans = KMeans(n_clusters = 5)
kmeans.fit(data.iloc[:,0:2])

# Menyimpan titik centroid ke dalam suatu variabel untuk divisualisasi nanti
centroids = kmeans.cluster_centers_
```

```python
# Visualisasi data yang sudah dilabel oleh kmeans dengan warna rainbow
plt.scatter(data.iloc[:,0], data.iloc[:,1], c = kmeans.labels_, cmap = 'rainbow')

# Visualisasi titik centroid dengan warna hitam dan opacity 50%
plt.scatter(centroids[:,0], centroids[:,1], s = 700, color = 'black', alpha = 0.5)
plt.xlabel("x")
plt.ylabel("y")
plt.title("Visualisasi clustering")
plt.show
```
